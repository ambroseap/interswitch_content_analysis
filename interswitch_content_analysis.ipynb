```python
# Convert text to lowercase and extract words
words = re.findall(r'\b\w+\b', most_used_words.lower())
hashtag = re.findall(r'\b\w+\b', most_used_hastagt.lower())
```

```python
# Count word occurrences
word_counts = Counter(words)
hashtag_counts = Counter(hashtag)
```
